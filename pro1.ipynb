```python
from deep_translator import GoogleTranslator
```

```python
languages = {
    "1": ("Tamil", "ta"),
    "2": ("Hindi", "hi"),
    "3": ("Telugu", "te"),
    "4": ("Malayalam", "ml"),
    "5": ("Kannada", "kn"),
    "6": ("French", "fr"),
    "7": ("German", "de"),
    "8": ("Spanish", "es"),
    "9": ("Japanese", "ja"),
    "10": ("Korean", "ko")
}

print("Select Target Language")
for key, value in languages.items():
    print(f"{key}. {value[0]}")

choice = input("Enter your choice: ")

text = input("Enter text to translate: ")

target_lang = languages[choice][1]

translated = GoogleTranslator(
    source='auto',
    target=target_lang
).translate(text)

print("\nTranslated Text:")
print(translated)
```
